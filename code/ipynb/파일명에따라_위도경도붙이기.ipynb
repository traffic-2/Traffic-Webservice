```python
import os

for file in globbed_files:
    frame = pd.read_excel(file)
    file_name = os.path.basename(file)     # os.path.basename(path) : 입력받은 경로의 기본 이름을 반환
    
    file_namena=file_name.split('_')
    file_gu=file_namena[0]
    file_jang=file_namena[1]
    
    file_jang=file_jang[:-4]
    
#     print('filename:',file_name)
    
#     if file_name.find('(') != -1:
#         idx = file_name.find('(')
#         file_name = file_name[:idx]
#     file_name = file_name.replace('.xls', '')
    
#     idx2 = file_name.find('_')
#     file_name = file_name[idx2 + 1:]
    
#     print('after filename:',file_name)
    
    
    
    frame['지점명'] = file_jang
    frame['지점명'][0] = '지점명'
    frame['지점명'][1] = ''
    
    find = latlng_df.loc[latlng_df['지점명'] == file_jang]
    lat = find['위도'].values[0]
    lng = find['경도'].values[0]
    
    frame['위도'] = lat
    frame['위도'][0] = '위도'
    frame['위도'][1] = ''
    
    frame['경도'] = lng
    frame['경도'][0] = '경도'
    frame['경도'][1] = ''
    frame['자치구'] = '강북구'
    frame.to_csv('C:/Users/cityo/Desktop/강서구/올빼미/양진이/%s.csv' %file_jang, encoding='utf-8-sig', header=False)
```
