```python
import pandas as pd
import numpy as np
```

```python
accident_list = ['안국역사거리', '광하문우체국사거리','종로3가역사거리', '이화사거리', '동묘앞역사거리']
traffic_list = ['율곡로(안국역)', '세종대로(시청역2)', '종로(종로3가역)', '대학로(한국방송통신대학교)', '종로(동묘앞역)']

data = {}
for i in range(len(accident_list)):
    data[traffic_list[i]] = accident_list[i]
    
data
```

### 테스트

```python
data_2017 = pd.read_csv('./교통량/2017/2017_1월.csv', encoding='utf-8')
data_2017 = data_2017[(data_2017['지점명'] == traffic_list[0]) | (data_2017['지점명'] == traffic_list[1]) | (data_2017['지점명'] == traffic_list[2]) | (data_2017['지점명'] == traffic_list[3]) | (data_2017['지점명'] == traffic_list[4])]
data_2017
```

```python
# 재구조화(unpivot)
df_unpivot = pd.melt(data_2017, id_vars=['년도','월','일','요일','구분','지점명'],value_vars=['0시', '1시', '2시', '3시', '4시',
       '5시', '6시', '7시', '8시', '9시', '10시', '11시', '12시', '13시', '14시', '15시',
       '16시', '17시', '18시', '19시', '20시', '21시', '22시', '23시'])
df_unpivot.rename(columns={'variable':'시간', 'value':'교통량'}, inplace=True)

df_unpivot
```

```python
# 일자별 시간별 정렬
df_unpivot= df_unpivot.sort_values(by=['년도','월','일','지점명','시간'],ascending=True)
df_unpivot
```

```python
# 시간열에 '시' 삭제 및 데이터 type 변경
df_unpivot['시간'] = df_unpivot['시간'].str.replace('시','')

df_unpivot['시간']= pd.to_numeric(df_unpivot['시간'],errors='coerce')
df_unpivot['교통량']= pd.to_numeric(df_unpivot['교통량'],errors='coerce')
```

```python
# 유입+유출의 평균으로 교통량 산정
df_final = pd.pivot_table(df_unpivot, values='교통량',index=['년도','월','일','요일','지점명','시간'],aggfunc='mean')
df_final.reset_index(inplace=True)
df_final
```

```python
# 교통량측정지점명에 매칭되는 사고다발지점 컬럼 추가
df_final = df_final.rename(columns={'지점명': "교통량측정지점명"})

df_final['사고다발지점'] = ''
for idx, val in df_final.iterrows():
    df_final['사고다발지점'][idx] = data[val['교통량측정지점명']]

df_final
```

```python
df_final = df_final[['년도','월','일','요일','시간','교통량측정지점명','사고다발지점','교통량']]
df_final
```

#### 1. 2017 교통량

```python
traffic_df = pd.DataFrame(columns=['년도','월','일','요일','시간','교통량측정지점명','사고다발지점','교통량'])
traffic_df
```

```python
for i in range(1, 13):
    # 교통량 데이터 파일 읽어옴
    data_df = pd.read_csv('./교통량/2017/2017_%d월.csv' % i, encoding='utf-8')
    
    # traffic_list에 해당하는 지점명만 추출
    data_df = data_df[(data_df['지점명'] == traffic_list[0]) | (data_df['지점명'] == traffic_list[1]) | (data_df['지점명'] == traffic_list[2]) | (data_df['지점명'] == traffic_list[3]) | (data_df['지점명'] == traffic_list[4])]
    
    # 재구조화(unpivot)
    df_unpivot = pd.melt(data_df, id_vars=['년도','월','일','요일','구분','지점명'],value_vars=['0시', '1시', '2시', '3시', '4시',
           '5시', '6시', '7시', '8시', '9시', '10시', '11시', '12시', '13시', '14시', '15시',
           '16시', '17시', '18시', '19시', '20시', '21시', '22시', '23시'])
    df_unpivot.rename(columns={'variable':'시간', 'value':'교통량'}, inplace=True)

    # 일자별 시간별 정렬
    df_unpivot= df_unpivot.sort_values(by=['년도','월','일','지점명','시간'],ascending=True)

    # 시간열에 '시' 삭제 및 데이터 type 변경
    df_unpivot['시간'] = df_unpivot['시간'].str.replace('시','')

    df_unpivot['시간']= pd.to_numeric(df_unpivot['시간'],errors='coerce')
    df_unpivot['교통량']= pd.to_numeric(df_unpivot['교통량'],errors='coerce')
    
    # 유입+유출의 평균으로 교통량 산정
    df_final = pd.pivot_table(df_unpivot, values='교통량',index=['년도','월','일','요일','지점명','시간'],aggfunc='mean')
    df_final.reset_index(inplace=True)
    
    # 교통량측정지점명에 매칭되는 사고다발지점 컬럼 추가
    df_final = df_final.rename(columns={'지점명': "교통량측정지점명"})

    df_final['사고다발지점'] = ''
    for idx, val in df_final.iterrows():
        df_final['사고다발지점'][idx] = data[val['교통량측정지점명']]

    df_final = df_final[['년도','월','일','요일','시간','교통량측정지점명','사고다발지점','교통량']]

    # traffic_df 에 추가
    traffic_df = traffic_df.append(df_final)
```
