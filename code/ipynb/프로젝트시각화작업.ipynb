```python
plt.figure(figsize=[18,8])
sns.countplot(data=df__gangdong, x='년도', hue='사고유형1')
plt.legend(loc='upper right', bbox_to_anchor=(1.0, 1))
plt.title('강동구 년도별 사고유형1 비교')
plt.xlabel('년도')
plt.ylabel('사고유형1')
plt.savefig('./image/강동구유형1.png')
```

```python
plt.figure(figsize=[18,8])
sns.countplot(data=df__gangdong, x='년도', hue='사고유형2')
plt.legend(loc='upper right', bbox_to_anchor=(1.0, 1))
plt.title('강동구 년도별 사고유형2 비교')
plt.xlabel('년도')
plt.ylabel('사고유형2')
plt.savefig('./image/강동구유형2.png')
```

```python
plt.figure(figsize=(30,8))
# fig,ax = plt.subplots(3,3)
colors=['red','green','blue','red','purple']


plt.subplot(1,3,1)
plt.title('3년간 사망자,중상자,경상자,부상자 분포')
df__gangdong[['사망자수','중상자수','경상자수','부상자수']].sum().plot.bar(rot=0,color=colors)


plt.subplot(1,3,2)
plt.title('3년간 사고등급 분포')
df__gangdong['사고등급'].value_counts().plot.bar(rot=0,color=colors)


plt.subplot(1,3,3)
plt.title('3년간 법규위반 분포')
df__gangdong['법규위반'].value_counts().head(4).plot.bar(rot=0,color=colors)
plt.savefig('./image/강동구_사망중상경상.png')
```
