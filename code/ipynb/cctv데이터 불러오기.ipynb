```python
import pandas as pd
import numpy as np

import requests
import json
import xmltodict
import sys
```

```python
# 공공데이터포털 `전국무인교통단속카메라표준데이터` 데이터 api 호출
url = 'http://api.data.go.kr/openapi/tn_pubr_public_unmanned_traffic_camera_api'
key = 'VaFzbCO33iSd7bRarAAGnJStsAFTn2wrjHHhHtXNeknK1mPX35OYZBWl96ST%2Fjsm30FxOaMox2KIo%2FEadUfsPQ%3D%3D'
url += '?serviceKey='
url += key

# 데이터 저장할 빈 데이터프레임 생성
df = pd.DataFrame(columns=['시도명', '시군구명', '도로노선명', '소재지도로명주소', '위도', '경도'])

for i in range(16):
    queryParams ={ 'pageNo' : i,
                   'numOfRows' : 100,
                   'type': 'xml',
                   'ctprvnNm' : '서울특별시'  }

    
    response = requests.get(url, params=queryParams)  
    print(i, response.status_code)
    
    if response.status_code == 200:
        xd = xmltodict.parse(response.text)        # XML형식의 데이터를 dict형식으로 변환
        xdj = json.dumps(xd, ensure_ascii=False)  # dict형식의 데이터를 json형식으로 변환
        xdjd = json.loads(xdj)                    # json형식의 데이터를 dict형식으로 변환

    # api로 호출된 데이터 가져오기
    data = xdjd['response']['body']['items']['item']

    # 필요한 데이터만 가져와서 데이터프레임에 행으로 추가
    for i in range(len(data)):
        temp_df = pd.DataFrame([[data[i]['ctprvnNm'], data[i]['signguNm'], data[i]['roadRouteNm'], data[i]['rdnmadr'], data[i]['latitude'], data[i]['longitude']]], columns=['시도명', '시군구명', '도로노선명', '소재지도로명주소', '위도', '경도'])
        df = df.append(temp_df)
```
